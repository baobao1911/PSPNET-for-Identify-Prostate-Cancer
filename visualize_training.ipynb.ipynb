```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt(r'D:\University\Semantic_Segmentation_for_Prostate_Cancer_Detection\Semantic_Segmentation_for_Prostate_Cancer_Detection\Training_result\Result_info\PSPNet.csv', \
                     delimiter=',', dtype=float)
train_loss = data[0].tolist()
train_mIou = data[1].tolist()
train_mAcc = data[2].tolist()
train_allAcc = data[3].tolist()
val_loss = data[4].tolist()
val_mIou = data[5].tolist()
val_mAcc = data[6].tolist()
val_allAcc = data[7].tolist()
```

```python
data = np.genfromtxt(r'D:\University\Semantic_Segmentation_for_Prostate_Cancer_Detection\Semantic_Segmentation_for_Prostate_Cancer_Detection\Training_result\Result_info\PSPNet_X.csv', \
                     delimiter=',', dtype=float)
train_loss1 = data[0].tolist()
train_mIou1 = data[1].tolist()
train_mAcc1 = data[2].tolist()
train_allAcc1 = data[3].tolist()
val_loss1 = data[4].tolist()
val_mIou1 = data[5].tolist()
val_mAcc1 = data[6].tolist()
val_allAcc1 = data[7].tolist()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
fig.tight_layout(pad=3)

ax[0].plot(val_loss, color='skyblue', label='PSPNet')
ax[0].plot(val_loss1, color='orange', label='PSPNet_CBAM_HDC2')
ax[0].set_xlim([-5, 185])
ax[0].legend()
ax[0].set_title('Comparison loss on validation')
ax[0].xaxis.set_label_text('num epochs')
ax[0].yaxis.set_label_text('Loss')

ax[1].plot(val_mIou, color='skyblue', label='PSPNet')
ax[1].plot(val_mIou1, color='orange', label='PSPNet_CBAM_HDC2')
ax[1].set_xlim([-5, 185])
ax[1].legend()
ax[1].set_title('Comparison mean IoU on validation')
ax[1].xaxis.set_label_text('num epochs')
ax[1].yaxis.set_label_text('mIoU')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
fig.tight_layout(pad=3)
#create subplots
ax[0, 0].plot(val_loss1, color='skyblue', label='Validation')
ax[0, 0].plot(train_loss1, color='orange', label='Train')
ax[0, 0].legend()
ax[0, 0].set_title('Train and validate loss')
ax[0, 0].xaxis.set_label_text('num epochs')
ax[0, 0].yaxis.set_label_text('Loss')


ax[0, 1].plot(train_mIou1, color='orange', label='Train')
ax[0, 1].plot(val_mIou1, color='skyblue', label='Validation')
ax[0, 1].legend()
ax[0, 1].set_title('Train and validate mean IoU ')
ax[0, 1].xaxis.set_label_text('num epochs')
ax[0, 1].yaxis.set_label_text('mean IoU')

ax[1, 1].plot(train_mAcc1, color='orange', label='Train')
ax[1, 1].plot(val_mAcc1, color='skyblue', label='Validation')
ax[1, 1].legend()
ax[1, 1].set_title('Model mean accurancy')
ax[1, 1].xaxis.set_label_text('num epochs')
ax[1, 1].yaxis.set_label_text('Mean Accurancy')

ax[1, 0].plot(train_allAcc1, color='orange', label='Train')
ax[1, 0].plot(val_allAcc1, color='skyblue', label='Validation')
ax[1, 0].legend()
ax[1, 0].set_title('Model accurancy')
ax[1, 0].xaxis.set_label_text('num epochs')
ax[1, 0].yaxis.set_label_text('Accurancy ')
plt.show()
```
